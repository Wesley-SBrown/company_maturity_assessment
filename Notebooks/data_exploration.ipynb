```python
import pandas as pd
import sys
import os
```

### Read Company Profile

```python
company_profile = pd.read_csv("../data/raw/Company_profile.csv")
```

```python
company_profile.head()
```

```python
for column in company_profile.columns: print(column)
```

```python
company_profile.info()
```

```python
# Get the absolute path of the scripts folder
scripts_path = os.path.abspath('../scripts')

# Add the scripts folder to the sys.path
sys.path.append(scripts_path)

from csv_filtering import filter_company_profile
```

```python
company_profile = filter_company_profile(company_profile).iloc[:-1]
company_profile.info()
```

```python
company_profile["Business Status"].unique()
```

```python
from csv_preprocessing import preprocess_company_profile
preprocessed_company_profile = preprocess_company_profile(company_profile)
preprocessed_company_profile.info()
```

```python
preprocessed_company_profile.to_csv("../data/cleaned/preprocessed_company_profile.csv")
```

### Read Deals

```python
deals = pd.read_csv("../data/raw/Deals.csv")
deals.info()
```

```python
print(", ".join(deals.columns))
```

```python
from csv_filtering import filter_deals
deals = filter_deals(deals)
deals.head()
```

```python
deals.info()
```

```python
from csv_preprocessing import preproces_deals
```

```python
preprocessed_deals = preproces_deals(deals)
```

```python
preprocessed_deals.to_csv('../data/cleaned/preprocessed_deals.csv')
```

```python
preprocessed_deals.info()
```

### Merge Company Profile and Deals on Company ID

```python
# Perform a one-to-many join (left) on Company Profile to attach each deal. 
company_and_deals = pd.merge(preprocessed_company_profile, preprocessed_deals, on='Company ID', how='left')\
                            .sort_values(by=['Company ID','Deal No.'])
company_and_deals.head()
```

```python
company_and_deals.to_csv('../data/cleaned/company_and_deals.csv')
```
